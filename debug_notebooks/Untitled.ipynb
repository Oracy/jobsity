```python
# 1
def drop_duplicated(df):
    # Trips with similar origin, destination, and time of day should be grouped together.
    data_drop_duplicated = df.drop_duplicates(subset = ["origin_coord", "destination_coord", "datetime"])
    return data_drop_duplicated


# 2
def convert_datetime_type(df):
    df['datetime'] = pd.to_datetime(df['datetime'])
    return df


# 3
def remove_characters(df):
    df['origin_coord'] = df['origin_coord'].str.replace('POINT \(', '', regex=True)
    df['origin_coord'] = df['origin_coord'].str.replace('\)', '', regex=True)
    df['destination_coord'] = df['destination_coord'].str.replace('POINT \(', '', regex=True)
    df['destination_coord'] = df['destination_coord'].str.replace('\)', '', regex=True)
    return df


# 4
def create_long_lat(df):
    df[['lat_origin', 'long_origin']] = df['origin_coord'].str.split(' ', 1, expand=True)
    df[['lat_destination', 'long_destination']] = df['destination_coord'].str.split(' ', 1, expand=True)
    return df


# 5
def remove_excedent_columns(df):
    df = df.drop(labels=['origin_coord', 'destination_coord'], axis=1)
    return df


# 6
def create_week_of_year_by_area(df):
    # Develop a way to obtain the weekly average number of trips for an area, defined by a bounding box (given by coordinates) or by a region.
    trips_by_area = df.groupby(["region", "weekofyear"]).size().reset_index(name='count')
    return trips_by_area
```

```python
def get_file_names(files_path, folder_name):
    """Get all file names that would be loaded, from specific folder and specific path.

    Args:
        files_path: File path that parquet are stored, to read it on local folders.
        folder_name: Folder name to find files.

    Returns:
        An array with all create tables. For example:
        ["acute_g_day_bw_all_days", "acute_g_day_bw_cons_days", "chronic_g_day_bw_to_t_pop", ...]
    """
    files_name = None
    for root, directories, files in os.walk(files_path):
        if folder_name in root:
            files_name = files
    files_name_regexp = [ name.replace(".sql", "") for name in files_name]
    return files_name_regexp
```

```python
def load_chunk_data(file_path, file_name, chunk_size):
    now = datetime.now()
    date_time = now.strftime("%m_%d_%Y")
    df = pd.read_csv(f"{file_path}/{file_name}", chunksize=chunk_size)
    src_path = f"{file_path}/{file_name}"
    dst_path = f"{file_path}/../done/{file_name}"
    part = 0
    for chunk in df:
        df_drop_duplicated = drop_duplicated(chunk)
        df_convert_datetime = convert_datetime_type(df_drop_duplicated)
        df_remove_characters = remove_characters(df_convert_datetime)
        df_create_long_lat = create_long_lat(df_remove_characters)
        df_remove_excedent_columns = remove_excedent_columns(df_create_long_lat)
        df_remove_excedent_columns.to_csv(f"{file_path}/chunks/{file_name}_{date_time}_{part}")
        part+=1
    shutil.move(src_path, dst_path)
```
